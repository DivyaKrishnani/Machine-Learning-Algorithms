```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Datasets/drug2.csv")
df
```

```python
df.shape
```

```python
df.describe()
```

```python
doseplussex = df['dose'] * df['sex']
doseplussex
```

```python
response = df['response']
response
```

```python
print(len(response.values))
print(response.values)
print(len(doseplussex.values))
print(doseplussex.values)
```

```python
Y = response
Y = np.array(Y).reshape(-1,1)
print(Y.shape)

X = doseplussex
X = np.array(X).reshape(-1,1)
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size = 0.30, random_state=0)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
ypred = lr.predict(xtest)
```

```python
import matplotlib.pyplot as plot
plot.scatter(X,Y)
plot.ylabel('response')
plot.xlabel('doseplussex')
plot.show()
```

```python
plot.scatter(xtrain, ytrain, color = 'red')
plot.plot(xtrain, lr.predict(xtrain), color = 'blue')
plot.title('Training Set')
plot.ylabel('response')
plot.xlabel('dose')
plot.show()
```

```python
plot.scatter(xtest, ytest, color = 'red')
plot.plot(xtrain, lr.predict(xtrain), color = 'blue')
plot.title('Test Set')
plot.ylabel('response')
plot.xlabel('dose')
plot.show()
```

```python
#Coeffients and Intercept
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_train_predict = lr.predict(xtrain)
rmse = (np.sqrt(mean_squared_error(ytrain, y_train_predict)))
r2 = r2_score(ytrain, y_train_predict)
```

```python
print(r2)
print(rmse)
```
