```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("Datasets/Wholesale_customers_data.csv")
df
```

```python
df2 = df.drop(['Channel', 'Region'], axis=1)
df2
```

```python
from sklearn.cluster import AgglomerativeClustering
 
```

```python
#linkage : {“ward”, “complete”, “average”, “single”}, optional (default=”ward”)
n_clusters=3
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
cluster.fit(df2)
```

```python
labels = cluster.labels_
labels
```

```python
clusters = {i: np.where(labels == i)[0] for i in range(n_clusters)}
clusters
```

```python
features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
features
```

```python
centers = []
for i in range(len(clusters)):
    cent_i = []
    #print(clusters[i])
    for j in features:
        cen = 0
        #print(df2.iloc[clusters[i]][j])
        #print(df2.iloc[clusters[2]][j])
        #print(len((df2.iloc[clusters[2]][j])))
        for k in df2.iloc[clusters[i]][j]:
            cen += k
        cen = cen/len(clusters[i])
        cent_i.append(cen)
    centers.append(cent_i)
centers   
```

## Total SSE, Sum of Squares Between(SSB), Sum of Squares Within(SSW)

```python
#SSW
sse_each_cluster = []
for i in clusters:
    sse_i = 0
    for j in range(len(clusters[i])):
        #print(clusters[i][j])
        k2=0
        for k in features:
            #print(df2.iloc[clusters[i][j]][k])
            #print()
            #print(centers[i][k2])
            sse_i += ((df2.iloc[clusters[i][j]][k]) - (centers[i][k2]))**2
            k2+=1
    sse_each_cluster.append(sse_i)
        
sse_each_cluster
```

```python
#Total Sum of Squares Within = Total SSE
SSW = 0
for i in sse_each_cluster:
    SSW += i
SSW
```

```python
#SSB
SSB = 0

for i in range(len(centers)):
    for j in range(i+1,len(centers)):
        for k in range(len(features)):
            #print(centers[i][k])
            #print(centers[j][k])
            SSB += ((centers[j][k] - centers[i][k])**2)


#print(centers[0])
#print(centers[1])
SSB
```

## Dendogram Plots

```python
import matplotlib.pyplot as plt
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(100, 20))  
plt.title("Dendograms with single linkage")  
dend = shc.dendrogram(shc.linkage(df2, method='single'))  
```

```python
plt.figure(figsize=(15, 8))  
plt.title("Dendograms with complete linkage")  
dend = shc.dendrogram(shc.linkage(df2, method='complete'))  
```

```python
plt.figure(figsize=(15, 8))  
plt.title("Dendograms with average linkage")  
dend = shc.dendrogram(shc.linkage(df2, method='average'))  
```

```python
plt.figure(figsize=(15, 8))  
plt.title("Dendograms")  
dend = shc.dendrogram(shc.linkage(df2, method='ward'))
```
