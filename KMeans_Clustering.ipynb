```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 0).fit(df2)
```

```python
labels = kmeans.labels_
labels
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
#No. of data entries in each cluster
num_clusters = [0,0,0,0,0]
for i in kmeans.labels_:
    num_clusters[i]+=1
num_clusters
```

```python
clusters = {i: np.where(kmeans.labels_ == i)[0] for i in range(kmeans.n_clusters)}
clusters
```

```python
#Total SSE
SSE = kmeans.inertia_
SSE
```

```python
sse = {}
for k in range(2, 21):
    kmeans1 = KMeans(n_clusters=k, max_iter=100).fit(df2)
    sse[k] = kmeans1.inertia_ 
    # Inertia: Sum of distances of samples to their closest cluster center
print(sse)
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```
